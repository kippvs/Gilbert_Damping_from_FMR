The Gilbert damping parameter, $\alpha$, can be obtained from the FMR linewidth if the saturation magnetization, $M_{eff}$, and magnetization angle, $\vartheta_M$, are known.  

See articles:  
[1] Maksymowicz, A. Z. & Leaver, K. D. Angular dependence of line width in nickel platelets. *J. Phys. F Met. Phys.* **3**, 1031–1038 (1973).  

[2] Mizukami, S., Ando, Y. & Miyazaki, T. Effect of spin diffusion on Gilbert damping for a very thin permalloy layer in Cu/permalloy/Cu/Pt films. *Phys. Rev. B* **66**, 104413 (2002).


```python
from __future__ import division
from sympy import *
init_printing(use_latex=True)
```

```python
# define some symbolic factors
factor_symbols = 'Delta, omega, gamma, theta_H, theta_M, 4piM_eff, alpha'
field_symbols = 'H_1, H_2, H_res, H_in, H_ex, H_pp'
delta_symbols = 'd4piMeff, dThetaH'

Delta, omega, gamma, thetaH, thetaM, Meff, alpha = symbols(factor_symbols, real=True)
H_1, H_2, H_res, H_in, H_ex, H_pp = symbols(field_symbols, real=True)
Delta4piMeff, DeltaThetaH = symbols(delta_symbols, real=True)

DeltaH_in = Delta*H_in
DeltaH_ex = Delta*H_ex
DeltaH_pp = Delta*H_pp
```

Once the microwave field and precessing magnetization are substituted into the Landau-Lifshitz-Gilbert equation, the FMR resonance condition can be obtained after considering the following relations (consequences of linear approximation):

```python
H1 = H_res*cos(thetaH - thetaM) - Meff*cos(2*thetaM)
Eq(H_1, H1)
```

```python
H2 = H_res*cos(thetaH - thetaM) - Meff*cos(thetaM)**2
Eq(H_2, H2)
```

```python
Larmor = sqrt(H1*H2)
Eq(omega/gamma, sqrt(H_1*H_2))
```

```python
Eq(omega/gamma, Larmor)
```

```python
# expand the terms under the radical
Eq(omega/gamma, expand(Larmor))
```

Angular dependence of magnetization can be obtained from:

```python
Hres_vs_thetaM = Eq(2*H_res*sin(thetaM - thetaH), Meff*sin(2*thetaM)); Hres_vs_thetaM
```

Where the FMR resonance center behaves as

```python
Hres =  solve(Hres_vs_thetaM, H_res)[0]
Eq(H_res, Hres)
```

For cases where the magnetization has low perpendicular anisotropy, the FMR FWHM linewidth is determined by two principle components: an intrinsic and extrinsic broadening.

```python
Eq(DeltaH_pp, (DeltaH_in + DeltaH_ex)/sqrt(3))
```

The intrinsic broadening is governed by the Gilber damping factor

```python
Eq(DeltaH_in, alpha*(H_1 + H_2)/abs(Derivative((omega/gamma), H_res)))
```

```python
deltaH_in = alpha*(H1 + H2)/abs(Derivative(Larmor, H_res))
Eq(DeltaH_in, deltaH_in)
```

```python
deltaH_in = deltaH_in.doit()
Eq(DeltaH_in, simplify(deltaH_in))
```

The extrinsic broadening follows the dispersion in magnetization magnitude and angle

```python
deltaH_ex = abs(Derivative(H_res, Meff))*Delta4piMeff + abs(Derivative(H_res, thetaH))*DeltaThetaH
Eq(DeltaH_ex, deltaH_ex)
```

```python
deltaH_ex = abs(Derivative(Hres, Meff))*Delta4piMeff + abs(Derivative(Hres, thetaH))*DeltaThetaH
deltaH_ex = deltaH_ex.doit()
Eq(DeltaH_ex, simplify(deltaH_ex))
```

Then the total FMR linewidth is

```python
deltaH_pp = (deltaH_in + deltaH_ex)/sqrt(3)
Eq(DeltaH_pp, deltaH_pp)
```

Now can solve for the Gilbert damping parameter, alpha

```python
D_in_expr = alpha*(H_1 + H_2)/abs(Derivative((omega/gamma), H_res))
D_ex_expr = abs(Derivative(H_res, Meff))*Delta4piMeff + abs(Derivative(H_res, thetaH))*DeltaThetaH
D_pp_expr = (D_in_expr + D_ex_expr)/sqrt(3)
Eq(DeltaH_pp, D_pp_expr)
```

```python
alpha_expr = solve(Eq(DeltaH_pp, D_pp_expr), alpha)[0]
Eq(alpha, alpha_expr)
```

```python
alpha_eqn = solve(Eq(DeltaH_pp, deltaH_pp), alpha)[0]
Eq(alpha, alpha_eqn)
```
